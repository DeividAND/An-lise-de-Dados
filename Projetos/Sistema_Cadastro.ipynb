```python
'''
Projeto 01 - Sistema de Cadastro de Funcionários

Descrição:
Você foi contratado para desenvolver um sistema básico de gerenciamento de
funcionários para uma pequena empresa. O sistema deve ser capaz de receber e
armazenar informações sobre os funcionários e realizar operações básicas de
manipulação desses dados.

Requisitos:
Utilize a função input() para cadastrar coletando dados do usuário, como nome,
cidade e estado em que reside, e-mail, idade, escolaridade, cargo e salário.

Funcionalidades do Sistema:
O sistema deve oferecer as seguintes funcionalidades:

•	Adicionar um novo funcionário ao sistema, solicitando informações como nome,
cargo, salário, e-mail, etc.

•	Listar todos os funcionários cadastrados.

•	Atualizar as informações de um funcionário existente com base em um
identificador único (como um número de identificação ou nome).
Exibir informações de um funcionário específico.

•	Permitir a remoção de um funcionário do sistema.

Observações Finais: Certifique-se de implementar um sistema de identificação
único para cada funcionário. Valide as entradas do usuário para garantir que as
operações sejam realizadas corretamente e sem erros.
'''
```

```python
print('Escolha a opção desejada')
print('Digite 1 para Cadastrar um funcionário')
print('Digite 2 para Consultar um funcionário')
print('Digite 3 para Atualizar um funcionário')
print('Digite 4 para Remover um funcionário')
menu = input('Informe o número da opção que deseja: ')
if 'funcionarios' in locals():
  pass
else:
  funcionarios = [[],[],[],[],[],[],[],[],[]]
if menu == '1':
  id1 = input('Informe o Id do Funcionário: ')
  if id1 not in funcionarios[0]:
    funcionarios[0].append(id1)
    funcionarios[1].append(input('Informe o nome do Funcionário: '))
    funcionarios[2].append(input('Informe a cidade do Funcionário: '))
    funcionarios[3].append(input('Informe o estado do Funcionário: '))
    funcionarios[4].append(input('Informe o email do Funcionário: '))
    funcionarios[5].append(input('Informe a idade do Funcionário: '))
    funcionarios[6].append(input('Informe a escolaridade do Funcionário: '))
    funcionarios[7].append(input('Informe o cargo do Funcionário: '))
    funcionarios[8].append(input('Informe o salário do Funcionário: '))
  else:
    print('Esse Id já foi cadastrado!!!')
elif menu == '2':
  id2 = input('Informe o Id do Funcionário: ')
  if id2 in funcionarios[0]:
    i2 = funcionarios[0].index(id2)
    print(f'''Informações do funcionário(a):
Nome:         {funcionarios[1][i2]}
Cidade:       {funcionarios[2][i2]}
Estado:       {funcionarios[3][i2]}
Email:        {funcionarios[4][i2]}
Idade:        {funcionarios[5][i2]} anos
Escolaridade  {funcionarios[6][i2]}
Cargo:        {funcionarios[7][i2]}
Salário:      R$ {funcionarios[8][i2]}
''')
  else:
    print('Funcionário não cadastrado!!!')
elif menu == '3':
  id3 = input('Informe o Id do Funcionário: ')
  if id3 in funcionarios[0]:
    i3 = funcionarios[0].index(id3)
    nome = input('Informe o nome do Funcionário: ')
    funcionarios[1][i3] = nome
    cidade = input('Informe a cidade do Funcionário: ')
    funcionarios[2][i3] = cidade
    estado = input('Informe o estado do Funcionário: ')
    funcionarios[3][i3] = estado
    email = input('Informe o email do Funcionário: ')
    funcionarios[4][i3] = email
    idade = input('Informe a idade do Funcionário: ')
    funcionarios[5][i3] = idade
    escolaridade = input('Informe a escolaridade do Funcionário: ')
    funcionarios[6][i3] = escolaridade
    cargo = input('Informe o cargo do Funcionário: ')
    funcionarios[7][i3] = cargo
    salario = input('Informe o salário do Funcionário: ')
    funcionarios[8][i3] = salario
  else:
    print('Funcionário não cadastrado!!!')
elif menu == '4':
  id4 = input('Informe o Id do Funcionário: ')
  if id4 in funcionarios[0]:
    i4 = funcionarios[0].index(id4)
    funcionarios[0].pop(i4)
    funcionarios[1].pop(i4)
    funcionarios[2].pop(i4)
    funcionarios[3].pop(i4)
    funcionarios[4].pop(i4)
    funcionarios[5].pop(i4)
    funcionarios[6].pop(i4)
    funcionarios[7].pop(i4)
    funcionarios[8].pop(i4)
  else:
    print('Funcionário não cadastrado!!!')
```
