```python
import pandas as pd
```

```python
# Criando uma lista para que possa ser lida como tabela em Pandas
nome = ['Douglas', 'Daniela', 'Pedro', 'Maria', 'Eduardo', 'Ester']
```

```python
# Criando uma tabela de coluna única em Pandas (Series)
serie = pd.Series(nome)
print(serie)
```

```python
# Criando uma tabela de várias colunas em Pandas (DataFrame)
nome = ['Douglas', 'Daniela', 'Pedro', 'Maria', 'Eduardo', 'Ester']
idade = [47, 9, 68, 67, 44, 39]
peso = [83, 35, 89, 65, 92, 67]
altura = [1.85, 1.39, 1.75, 1.68, 1.82, 1.72]
df = pd.DataFrame(zip(nome,idade,peso,altura),columns=['nome','idade','peso','altura'])
display(df)
```

```python
medidas = df.describe()
display(medidas)
```

```python
display(df)
```

```python
# Mudando o valor de uma célula específica
df.iat[0,1] = 48
display(df)
```

```python
# Criando uma nova coluna com valores pré-definidos
df['sexo'] = ['Masculino','Feminino','Masculino','Feminino','Masculino','Feminino']
display(df)
```

```python
df['peso'] = [85,37,92,63,95,65]
display(df)
```

```python
df['imc'] = round(df['peso']/df['altura']**2,1)
display(df)
```

```python
# renomeando colunas de um df
df.rename(columns={'peso':'massa','sexo':'genero'},inplace=True)
display(df)
```

```python
# mudando os valores de múltiplas células ao mesmo tempo
df['genero'] = df['genero'].map({'Masculino':'M','Feminino':'F'})
display(df)
```

```python
# Ordenando o df de acordo com uma coluna específicade forma crescente
df = df.sort_values('idade')
display(df)
```

```python
# Ordenando o df de acordo com uma coluna específicade forma decrescente
df = df.sort_values('idade',ascending=False)
display(df)
```

```python
# Ordenando o df por múltiplas colunas
df = df.sort_values(['idade','imc'])
display(df)
```

```python
# reoordenando o índice do df
df.reset_index(inplace=True)
display(df)
```

```python
# Excluindo uma coluna
df.drop('index',axis=1,inplace=True)
display(df)
```

```python
# Excluindo linhas de uma df
df.drop(3,axis=0,inplace=True)
display(df)
```

```python
# reoordenando o índice do df
df.reset_index(drop=True,inplace=True)
display(df)
```
