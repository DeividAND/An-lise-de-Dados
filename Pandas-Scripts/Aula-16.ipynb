```python
# Criando uma tabela de várias colunas em Pandas (DataFrame)
nome = ['Douglas', 'Daniela', 'Pedro', 'Maria', 'Eduardo', 'Ester']
idade = [47, 9, 68, 67, 44, 39]
peso = [83, 35, 89, 65, 92, 67]
altura = [1.85, 1.39, 1.75, 1.68, 1.82, 1.72]
df = pd.DataFrame(zip(nome,idade,peso,altura),columns=['nome','idade','peso','altura'])
display(df)
```

```python
medidas = df.describe()
display(medidas)
```

```python
# Criando uma nova coluna com valores pré-definidos
df['sexo'] = ['Masculino','Feminino','Masculino','Feminino','Masculino','Feminino']
display(df)
```

```python
df['imc'] = round(df['peso']/df['altura']**2,1)
display(df)
```

```python
# renomeando colunas de um df
df.rename(columns={'peso':'massa','sexo':'genero'},inplace=True)
display(df)
```

```python
# mudando os valores de múltiplas células ao mesmo tempo
df['genero'] = df['genero'].map({'Masculino':'M','Feminino':'F'})
display(df)
```

```python
df = df.sort_values(['idade', 'imc'])
display(df)
```

```python
# reoordenando o índice do df
df.reset_index(drop=True,inplace=True)
display(df)
```

```python
'''
No jargão de análise de dados, as variáveis categóricas ou qualitativas, são
chamadas de dimensões e as variáveis quantitativas são chamadas de métricas,
para as métricas nós podemos calcular as medidas descritivas, como contagem,
média, mediana, desvio padrão e etc.
'''
```

```python
'''
Calcular a média dos salários
'''
df['salario'].mean()
```

```python
'''
Calcular a media do salário
'''
df['salario'].median()
```

```python
'''
Calcular a média e a mediana dos salários
'''
df.agg({'salario':['mean','median']})
```

```python
'''
Calcular a média dos salários e das idades
'''
df[['salario','idade']].mean()
```

```python
'''
Calcular a mediana dos salários e das idades
'''
df[['salario','idade']].median()
```

```python
'''
Calcular a média e a mediana dos salários e das idades
'''
df.agg({'salario':['mean','median'], 'idade':['mean','median']})
```

```python
'''
A biblioteca Pandas oferece poderosas funções para agrupamento e agregação de
dados. Entender a diferença entre esses dois conceitos, bem como sua relação com
métricas e dimensões, é fundamental para realizar análises eficazes.


Agrupamento (Grouping):
Agrupamento é o processo de dividir um conjunto de dados em grupos menores com
base em alguma característica ou coluna(s) específica(s). No Pandas, isso é
feito usando o método groupby(). O agrupamento em si não modifica os dados, mas
prepara para operações de agregação ou transformação.

Agregação (Aggregation):
Agregação é o processo de aplicar uma função estatística a cada grupo criado
pelo método groupby(). Exemplos de funções de agregação incluem soma, média,
contagem, etc. O resultado da agregação é um DataFrame reduzido, onde os dados
foram combinados conforme a operação especificada.

Relação com Métricas e Dimensões
Dimensões: As dimensões são categorias ou atributos pelos quais os dados são
agrupados, geralmente são as variáveis catégoricas ou qualitativas.

Métricas: As métricas são os valores numéricos sobre os quais as operações de
agregação são realizadas, para usamos as variáveis quantitativas (numéricas).
'''
```

```python
'''
Quando você agrupa um DataFrame usando groupby(), o objeto resultante é um
DataFrameGroupBy, que não exibe os dados diretamente. Para visualizar os dados
agrupados, você precisa aplicar uma operação de agregação ao objeto agrupado.
'''
departamento = df.groupby('departamento')
display(departamento)
```

```python
'''
a função lambda x: x retorna os próprios dados agrupados, permitindo a
visualização direta dos grupos formados.
'''
departamento = df.groupby('departamento').apply(lambda x: x)
display(departamento)
```

```python
'''
Agora vamos contar quantos funcionários temos em cada departamento para cada
atributo

'''
depto_cont = df.groupby('departamento').count()
display(depto_cont)
```

```python
'''
Agora vamos contar quantos nomes temos em cada departamento
'''
depto_cont = df[['departamento','nome']].groupby('departamento').count()
display(depto_cont)
```

```python
'''
Agora vamos contar quantos nomes e quantos cargos temos em cada departamento

'''
depto_cont = df[['departamento','nome','cargo']].groupby('departamento').count()
display(depto_cont)
```

```python
'''
Agora vamos construir uma visão agrupada por departamento e agregando pela média
os salários
'''
round(df[['departamento','salario']].groupby('departamento').mean(),2)
```

```python
'''
Agora vamos construir uma visão agrupada por departamento e agregando pela média
e pela mediana os salários
'''
df.groupby('departamento').agg({'salario':['mean','median']})
```

```python
'''
Agora vamos construir uma visão agrupada por departamento e cargo,  agregando
pela média dos salários
'''
df[['departamento', 'cargo','salario']].groupby(['departamento','cargo']).mean()
```

```python
'''
Agora vamos construir uma visão agrupada por departamento e cargo,  agregando
pela média e mediana dos salários
'''
df.groupby(['departamento', 'cargo']).agg({'salario':['mean','median']})
```
